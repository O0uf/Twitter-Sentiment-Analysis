### Missing Data & Duplicated Data

```python
def missing_value_of_data(data):
    total=data.isnull().sum().sort_values(ascending=False)
    percentage=round(total/data.shape[0]*100,2)
    return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])
```

```python
missing_value_of_data(train)
```

```python
train = train.dropna()
```

```python
def duplicated_values_data(data):
    dup=[]
    columns=data.columns
    for i in data.columns:
        dup.append(sum(data[i].duplicated()))
    return pd.concat([pd.Series(columns),pd.Series(dup)],axis=1,keys=['Columns','Duplicate count'])
```

```python
duplicated_values_data(train).style.background_gradient(cmap='Purples')
```

```python
def col_dist_overview(data,col):
    total = data.loc[:,col].value_counts()
    freq = data.loc[:,col].value_counts(normalize=True)
    return pd.concat([total, freq],axis=1,keys=["Count","Frequency"]
                    ).style.background_gradient(cmap='Purples')
```

## Preprocessing Text 
Usually the steps includes 

1. Scrape text from raw documents
2. remove punctuation
3. lower case
4. tokenize & remove stop word 
5. lemmatize (lemma or stem)

In our case step1 is unnecessary.

```python
def twit_preproc(df,column,now):
    """Preprocessing for df[column]
        process involved: 
            - remove punctuation
            - lower case
            - tokenize & remove stop word 
            - lemmatize (lemma or stem)
    """

    def clean_text(text):
        '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
        text = str(text).lower()
        text = re.sub('\[.*?\]', '', text) # remove text in brackets
        text = re.sub('https?://\S+|www\.\S+', '', text) #remove link
        text = re.sub('<.*?>+', '', text) 
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('\n', '', text)
        text = re.sub('\w*\d\w*', '', text)
        return text
    df[now]= df[column].apply(lambda x:clean_text(x))
    
    # Tokenize & to lower case
    tokenizer = RegexpTokenizer(r'\w+')
    df[now] = df[now].apply(lambda x:tokenizer.tokenize(x))

    def remove_stopword(x):
        return [y for y in x if y not in stopwords.words('english')]
    df[now] = df[now].apply(lambda x:remove_stopword(x))
    
    # lemmatize and join the words
    lemmatizer = WordNetLemmatizer()
    def sentence_lemmatize(text):
        return " ".join([lemmatizer.lemmatize(x) for x in text])
    df[now] = df[now].apply(lambda text:sentence_lemmatize(text))
    
    return df
```

```python
# preprocess text from training set
train = twit_preproc(train,'text','cleaned')
```

```python
# proprocess testing set
test = twit_preproc(test,'text','cleaned')
```

### Maps sentiment to numbers

```python
sentiment_map = {
    'negative' : 0,
    'neutral' : 1,
    'positive' : 2
}
train['sentiment'] = train['sentiment'].map(sentiment_map)
test['sentiment'] = test['sentiment'].map(sentiment_map)
```

## Feature Extraction 
Here we use **bag of words** approach, which ignores the grammar of words and throw them into a vocabulary list. 

Then we use [**Count Vectorizer**](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) to vectorize the text input so that we can run ML models on it. It simply counts the number of occurence of a word in a sentence. 


```python
from sklearn.feature_extraction.text import CountVectorizer

bow_vectorizer = CountVectorizer(max_df=0.90, # ignores word with freq > .9
                                 min_df=2, # ignores word with coun < 2
                                 max_features=800, # size of the dictonary
                                 stop_words='english')
#bow = bow_vectorizer.fit_transform(series)
```
